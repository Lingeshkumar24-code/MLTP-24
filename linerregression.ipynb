```python
def my_slope(a, b):
    xi, yi = a
    xj, yj = b
    return (yj - yi) / (xj - xi)
```

```python
def my__slope(a,b):
    dy = a[1] - b[1]
    dx = a[0] - b[0]
    return dy/dx
```

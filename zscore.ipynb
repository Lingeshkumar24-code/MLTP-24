```python
import numpy as np
import math
```

```python
df = np.random.randint(2,100,size=5)
df
```

zscore = (x-mean)/sd , sd = root(sum((x-mean)**2)/n)

```python
mean = np.mean(df)
sum = 0
for i in range(len(df)):
	sum += (df[i] - mean)**2
sd = math.sqrt(sum/len(df))

for i in range(len(df)):
	zscore = (df[i] - mean) / sd
	print(f"Z-score of {df[i]}: {zscore}")
	if(zscore > 1 or zscore < -1):
		print(f"{df[i]} is a outlier")
```

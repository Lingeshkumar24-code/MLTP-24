```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

For continous numerical data we use Line chart


```python
x = [1, 2, 3, 4]              
y = (10.5, 20.2, 30.1, 40)    

plt.plot(x, y)
plt.show()
```

For Catgorial data we use BAR CHART


```python
categories = ["HR", "IT", "Sales"]   
values = [10, 25, 15]               

plt.bar(categories, values)
plt.show()
```

used for only one numerical column we use HISTOGRAM

```python
data = [22, 30, 40, 50, 35, 40]   

plt.hist(data, bins=4)
plt.show()
```

FOR outliers and spread we use BOX PLOT

```python
data=[20,25,30,35,40,200,400]
plt.boxplot(data)
plt.show()
```

Relataionship b/w two varibles we use SCATTER PLOT


```python
x = [1, 2, 3, 4]    
y = [5, 6, 7, 8] 

plt.scatter(x, y)
plt.show()
```
